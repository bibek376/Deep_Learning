## Tensorflow Loss Value



```python
import numpy as np
y_predict=np.array([1,0,1,1,0])
y_true=np.array([0.30,0.5,1,0,0.9])
```

```python
for (i,j) in zip(y_predict,y_true):
    print(i,j)
```

```python
total_diff=[]
for (i,j) in zip(y_predict,y_true):
    total_diff.append(abs(i-j))
print(total_diff)
```

```python
def MAE(y_predict,y_true):
    total_diff=[]
    for (i,j) in zip(y_predict,y_true):
        total_diff.append(abs(i-j))
    print("Total Error ",sum(total_diff))
    print("\nMAE")
    return sum(total_diff)/len(y_predict)
MAE([1,0,1,1,0],[0.30,0.5,1,0,0.9])
```

### Binary Cross Entropy or Log Loss

```python
y_predict_new=[max(i,epsilon) for i in y_predict]
y_predict_new
```

```python
y_predict_new=[min(i,1-epsilon) for i in y_predict_new]
y_predict_new
```

```python
np.log(y_predict_new)
```

```python
-np.mean(y_true*np.log(y_predict_new)+(1-y_true)*np.log(y_predict_new))
```
